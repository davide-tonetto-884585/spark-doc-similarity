```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.lines as mlines
```

# Results
Let's show the results of the algorithm obtained by running the code with different parameters.

## Speedup

```python
# load csv file into dataframe
simhash_time = pd.read_csv('./result/simhash_time.csv', header=None)
group_blocks_time = pd.read_csv('./result/group_blocks_time.csv', header=None)
similar_docs_time = pd.read_csv('./result/similar_docs_time.csv', header=None)

# rename columns
simhash_time.columns = ['n_cores', 'm', 'p', 'time'] 
group_blocks_time.columns = ['n_cores', 'm', 'p', 'time']
similar_docs_time.columns = ['n_cores', 'm', 'p', 'time']

max_threads = simhash_time['n_cores'].max()

# plot a graph foreach m and draw a line foreach p
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(25, 8))
for i, m in enumerate(simhash_time['m'].unique()):        
    ax[i].set_title(f'm = {m}')
    for p in simhash_time['p'].unique():
        seq_time = simhash_time[(simhash_time['m'] == m) & (simhash_time['p'] == p) & (simhash_time['n_cores'] == 1)]['time'].values[0]
        ax[i].plot(simhash_time[(simhash_time['m'] == m) & (simhash_time['p'] == p)]['n_cores'], seq_time / simhash_time[(simhash_time['m'] == m) & (simhash_time['p'] == p)]['time'], label=f'p = {p}')
    ax[i].set_xlabel('n_cores')
    ax[i].set_ylabel('SpeedUp')
    ax[i].legend()
    ax[i].grid()
    
    line = mlines.Line2D([0, max_threads], [0, max_threads], color='red', linestyle="dashed", label="Linear speedup")
    ax[i].add_line(line)

    ax[i].fill_between([x for x in range(0, max_threads + 1)],
                     [1] * (max_threads + 1),
                     alpha=0.3, color="red", label="No improvement zone")

    ax[i].fill_between([x for x in range(0, max_threads + 1)],
                     [x for x in range(0, max_threads + 1)],
                     alpha=0.1, color="blue", label="Sublinear speedup zone")

    ax[i].fill_between([x for x in range(0, max_threads + 1)],
                     [x for x in range(0, max_threads + 1)],
                     max_threads, alpha=0.1, color="green", label="Super linear speedup zone")
    

fig.suptitle('Simhash time speedup')
```

Total time is the sum of the time spent in each step of the algorithm.

```python

# plot a graph foreach m and draw a line foreach p
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(25, 8))
for i, m in enumerate(simhash_time['m'].unique()):        
    ax[i].set_title(f'm = {m}')
    for p in simhash_time['p'].unique():
        seq_time = simhash_time[(simhash_time['m'] == m) & (simhash_time['p'] == p) & (simhash_time['n_cores'] == 1)]['time'].values[0] \
            + group_blocks_time[(group_blocks_time['m'] == m) & (group_blocks_time['p'] == p) & (group_blocks_time['n_cores'] == 1)]['time'].values[0] \
            + similar_docs_time[(similar_docs_time['m'] == m) & (similar_docs_time['p'] == p) & (similar_docs_time['n_cores'] == 1)]['time'].values[0]
        total_time = simhash_time[(simhash_time['m'] == m) & (simhash_time['p'] == p)]['time'] \
            + group_blocks_time[(group_blocks_time['m'] == m) & (group_blocks_time['p'] == p)]['time'] \
            + similar_docs_time[(similar_docs_time['m'] == m) & (similar_docs_time['p'] == p)]['time']
        ax[i].plot(simhash_time[(simhash_time['m'] == m) & (simhash_time['p'] == p)]['n_cores'], seq_time / total_time, label=f'p = {p}')
    ax[i].set_xlabel('n_cores')
    ax[i].set_ylabel('SpeedUp')
    ax[i].legend()
    ax[i].grid()
    
    line = mlines.Line2D([0, max_threads], [0, max_threads], color='red', linestyle="dashed", label="Linear speedup")
    ax[i].add_line(line)

    ax[i].fill_between([x for x in range(0, max_threads + 1)],
                     [1] * (max_threads + 1),
                     alpha=0.3, color="red", label="No improvement zone")

    ax[i].fill_between([x for x in range(0, max_threads + 1)],
                     [x for x in range(0, max_threads + 1)],
                     alpha=0.1, color="blue", label="Sublinear speedup zone")

    ax[i].fill_between([x for x in range(0, max_threads + 1)],
                     [x for x in range(0, max_threads + 1)],
                     max_threads, alpha=0.1, color="green", label="Super linear speedup zone")
    

fig.suptitle('Total time speedup')
```

## Number of documents to compare
Now we plot the % of comparisons between documents needed to be done after grouping blocks with respect to the total number of comparisons needed without grouping blocks. 

```python
comparisons = pd.read_csv('./result/comparisons.csv', header=None)
comparisons.columns = ['n_cores', 'm', 'p', 'comparisons_ratio']

# bar plot withm mean comparisons % of each p for each m
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(25, 8))
for i, m in enumerate(comparisons['m'].unique()):        
    ax[i].set_title(f'm = {m}')
    ax[i].set_xticks(comparisons['p'].unique())
    
    for p in comparisons['p'].unique():
        ax[i].bar(p, comparisons[(comparisons['m'] == m) & (comparisons['p'] == p)]['comparisons_ratio'].mean() * 100, label=f'p = {p}', width=10)
    
    ax[i].set_xlabel('p')
    ax[i].set_ylabel('Comparisons %')
    ax[i].legend()

fig.suptitle('Comparisons % over total comparisons')
```

## Number of similar pairs
The following plot shows the number of similar pairs found by the algorithm for each m and p. 

```python
num_similar_docs = pd.read_csv('./result/num_similar_docs.csv', header=None)
num_similar_docs.columns = ['n_cores', 'm', 'p', 'num_similar_docs']

# mean and std of num_similar_docs for each m and p
num_similar_docs_mean = num_similar_docs.groupby(['m', 'p']).mean()
num_similar_docs_std = num_similar_docs.groupby(['m', 'p']).std()

# bar plot withm mean num_similar_docs of each p for each m
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(25, 8))
for i, m in enumerate(num_similar_docs['m'].unique()):        
    ax[i].set_title(f'm = {m}')
    ax[i].set_xticks(num_similar_docs['p'].unique())
    ax[i].set_ylim(0, num_similar_docs['num_similar_docs'].max())
    
    for p in num_similar_docs['p'].unique():
        ax[i].bar(p, num_similar_docs_mean.loc[(m, p)]['num_similar_docs'], yerr=num_similar_docs_std.loc[(m, p)]['num_similar_docs'], label=f'p = {p}', width=10)
    
    ax[i].set_xlabel('p')
    ax[i].set_ylabel('Num similar docs')
    ax[i].legend()
    
fig.suptitle('Number of similar pairs')
```
